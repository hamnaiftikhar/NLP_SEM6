```python
from nltk.corpus import twitter_samples
```

```python
# Load dataset
pos_tweets = twitter_samples.strings('positive_tweets.json')
neg_tweets = twitter_samples.strings('negative_tweets.json')
neutral_tweets = twitter_samples.strings('tweets.20150430-223406.json')

tweets = pd.DataFrame({'text': pos_tweets + neg_tweets + neutral_tweets,'sentiment': ['positive']*len(pos_tweets) + ['negative']*len(neg_tweets) + ['neutral']*len(neutral_tweets)})
```

##### Task 1.1 Count the number of positive, negative, and neutral tweets

```python
print("Number of Positive Tweets:", len(tweets[tweets['sentiment']=='positive']))
print("Number of Negative Tweets:", len(tweets[tweets['sentiment']=='negative']))
print("Number of Neutral Tweets:", len(tweets[tweets['sentiment']=='neutral']))
```

##### Task 1.2  Analyze  the  frequency  of  words  in  the  positive,  negative  and  neutral tweets  and  print  the  10  most  common words

```python

for sentiment in ['positive', 'negative', 'neutral']:
    words = tweets[tweets['sentiment'] == sentiment]['text'].str.split().explode().value_counts().head(10)
    print(f"Most common words in {sentiment} tweets:\n{words}\n")
    
#explode() method to transform the list of words for each tweet into separate rows, and then applies the value_counts() method to count the frequency of each word
```

##### Task 1.3 Analyze the frequency of hashtags in the positive, negative and neutral tweets and print the 10 most common hashtags

```python
for sentiment in ['positive', 'negative', 'neutral']:
    hashtags = tweets[tweets['sentiment'] == sentiment]['text'].str.findall(r'#\w+').explode().value_counts().head(10)
    print(f"Most common hashtags in {sentiment} tweets:\n{hashtags}\n")

#findall() method to extract all words that start with the # character from the text
```

##### Task 1.4 Analyze the frequency of retweets in the positive, negative and neutral tweets and print the 10 most common retweet

```python
all_tweets = pos_tweets + neg_tweets + neutral_tweets
```

```python
retweets = []
sentiments = []

for tweet in pos_tweets:
    retweets.append(tweet.count('RT'))
    sentiments.append('positive')

for tweet in neg_tweets:
    retweets.append(tweet.count('RT'))
    sentiments.append('negative')

for tweet in neutral_tweets:
    retweets.append(tweet.count('RT'))
    sentiments.append('neutral')
```

```python
retweet_counts = {'positive': {}, 'negative': {}, 'neutral': {}} #a dictionary to store the retweet counts for each sentiment

for tweet, sentiment in zip(all_tweets, sentiments):
    retweets = tweet.count('RT')
    if retweets not in retweet_counts[sentiment]:
        retweet_counts[sentiment][retweets] = 0
    retweet_counts[sentiment][retweets] += 1

for sentiment in ['positive', 'negative', 'neutral']: 
    print(f"Most common retweet counts in {sentiment} tweets:")
    for count, freq in sorted(retweet_counts[sentiment].items(), key=lambda x: x[1], reverse=True)[:10]: #loop to calculate top 10 retweets
        print(f"{count}: {freq}")
    print()
```
